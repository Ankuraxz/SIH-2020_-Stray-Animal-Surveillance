```
import cv2
import matplotlib.pyplot as plt
import os
import numpy as np
```

```
%cd /content/drive/My Drive
path= os.path.join(("/content/drive/My Drive/COWS"))#path here
print(path)
```

```
os.listdir(path) #should show name of all images
```

```
img_list=[]

for im in os.listdir(path):
        # print(os.path.join(path,im))
        # print (im)
        img = cv2.imread(os.path.join(path,im),cv2.IMREAD_UNCHANGED )# reading in grayscale, can be changed to RGB, RGBA, BGR
        #cv2.IMREAD_GRAYSCALE #and later
        # cv2.COLOR_BGR2RGB
        img_list.append(img)
        plt.imshow(img)
print(type(img))

        
```

# Resizing and Recolouring

```
for ix in range(0,len(img_list)):
  img_list[ix] = cv2.cvtColor(img_list[ix],cv2.COLOR_BGR2RGB)# Colours can be altered here
  img_list[ix]=cv2.resize(img_list[ix],dsize=(40,40))# change d_size here(height,width)
  #img_list[ix] = cv2.cvtColor(img_list[ix],cv.COLOR_BGR2RGB)# Colours can also be altered here
```

# SAVING IMAGES

```
# AS JPEG
%cd /content/drive/My Drive

directory = r'/content/drive/My Drive/COWS' #will be saved here
os.chdir(directory) 
from PIL import Image
for im in range(0,len(img_list)):
  im_str =str(im)
  
  str_name = "img"+im_str+".jpeg"
  #print(str_name)
  
  img1 = Image.fromarray(img_list[im])
  plt.imshow(img1)
  img1.save(str_name)
  
  print(str_name) # will be saved in Sample_data in google drive
```
